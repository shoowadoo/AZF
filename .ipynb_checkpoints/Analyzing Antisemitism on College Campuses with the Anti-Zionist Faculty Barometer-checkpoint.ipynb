## Data collection and preparation

### Webscraping data

We will be using Selenium and BeautifulSoup for scraping our data from the webpage.

```python
# uncomment below if needed to install 
#!pip install selenium beautifulsoup4
```

We will need the following imports for this purpose.

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from bs4 import BeautifulSoup
```

We want to set up a `webdriver` for us to control a Chrome browser. We use this webdriver to tell Chrome to visit the AMCHA Initiative webpage where the Anti-Zionist Faculty Barometer is located. The web address we provide our webdriver also contains special instructions to display 1000 entries per page and to start with the first page of the results. This allows us to acquire all of the webpage's content at once.

```python
driver = webdriver.Chrome()

driver.get('https://amchainitiative.org/azf-barometer/?ppwp=1#anti-zionist-faculty-barometer/?ppwp=1&view_433_per_page=1000&view_433_page=1')
```

Next, we retrieve the webpage's content and hand it over to BeautifulSoup for processing, and let it know we will be working with HTML data. We can then print the parsed content, showing the structure and data of the webpage.

Note: Please make sure the webdriver is fully loaded before running the next cell, or we will be missing content.

```python
page_source = driver.page_source

soup = BeautifulSoup(page_source, 'html')

#print(soup)
```

We now have access to the entire contents of the webpage and can extract the information we want. We are particularly interested in the tabular data, but first let's display all the rich text data from the webpage. Of course, we can read this information on the webpage itself, but we can also explore it here. We can use the `find_all()` method to extract all sections of the webpage that have rich text.

```python
info = soup.find_all('div', class_="fl-rich-text")
info[0]
```

We have now extracted all sections of the webpage that have the tag `<div class="fl-rich-text">`. Each section is contained in a list, and we display the first one above. Notice how the first line is the tag we told `soup` to find. Also, notice how the second line contains textual information after the tag `<p><em>`. This is the information we want, but it would be nice to see it displayed without all the HTML, just like we do when we go to the actual webpage. To do this, we just have to move through each section in our list one by one and tell it that we want just the rich text.

```python
for p in info:
    print(p.text)
```

We have now extracted and displayed all the rich text from the webpage. Of particular interest for our purposes is the information about the different rankings and what they mean. We will need to know this information to interpret the tabular data later.

Let's now extract the tabular data, which consists of rows and columns, from the webpage.

```python
table = soup.find('div', class_='kn-table-wrapper')
table
```

We can display all the column header names like this: 

```python
# table header column names 
for th in table.find_all('th'):
    print(th.text.strip())
```

But in order to be able to do statistical analysis with the data from the webpage, we need to structure our data in a way that can be easily processed and analyzed, such as by cleaning, organizing, and converting it into numerical or categorical formats suitable for statistical methods.

We can do this by creating our own table or dataframe using the Pandas library.

### Setting up the dataframe

We begin by setting up the dataframe with our column header names. First, let's put all column header names into a list. Then, we can import Pandas and easily add the columns we want to a Pandas dataframe.

```python
column_titles = []

for th in table.find_all('th'):
    column_titles.append(th.text.strip())

column_titles
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(columns=column_titles[:-1])
df
```

We've added every column but the last to our dataframe and displayed our dataframe `df`, which currently just has our header names. The next thing we want to do is to scrape each row of data and then add each row to our dataframe.

### Scraping each row of data

In the HTML we scraped from the webpage, each row in our table begins with a `<tr>` tag for 'table row', and we can find them all in our `table` object, again by using `find_all()`. However, we will need to skip the first `<tr>` which is the row of header names we've already added to our dataframe `df`. After finding and storing all `<tr>` tags in a `data` object, we display the first row. Notice the `<tr>` tag in the first line.

```python
data = table.find_all('tr')[1:]
data[0]
```

Again, we can display just the text (for the first row) without all that distracting HTML. 

```python
row_data = data[0].find_all('td')
for data_point in row_data[:-1]:
    print(data_point.text.strip())
```

Above are the data points for our first row of data in our table, one data point for each one of our columns. We will iterate through each row of our `data` object, extract all data points of each row, and add each data point of a row to a list of `school_data`. We will then add this list of school data to a master list of all schools. Afterwards, we can inspect the first 5 schools in our master list. 

```python
schools = []
for row in data: 
    row_data = row.find_all('td')
    school_data = [data_point.text.strip() for data_point in row_data[:-1]]
    schools.append(school_data)
schools[:5]
```

We can determine the number of schools in our master list by getting the length of the master list.

```python
num_of_schools = len(schools)
print("Number of schools: " + str(num_of_schools))
```

We are now ready to add all this data to our dataframe.

### Construct dataframe

Let's examine our dataframe so far, which currently has only the column header names.

```python
df
```

Each school in our master list of `schools` will be a row of our dataframe. We move school by school in our master list, adding each school's row of data to our dataframe. Once complete, we inspect the the first 5 rows of our dataframe as well as the number of rows in our dataframe, which should be the same as the number of schools in our master list.

```python
for school in schools:
    length = len(df)

    df.loc[length] = school
df.head()
```

```python
print("Number of rows in the dataframe: " + str(len(df)))
```

Our dataframe now consists of all 726 school's data in our master list. The first five rows all have a school ranking of '5 - Extreme.' We can inspect the last 5 rows of our dataframe as well.

```python
df.tail()
```

The last five rows all have a school ranking of '0 - Negligible.' It appears our dataframe is in descending order, beginning with schools ranked highest in anti-Zionist faculty.

Let's get some info about our newly constructed dataframe. 

```python
df.info()
```

From the above, we can see that our dataframe has 6 columns (to be expected) and that each column has 726 `non_null` values. This means that all rows have a data point for each one of the columns, and there are no data missing from the dataframe. Under the column `Dtype` above (not in our dataframe), the data type of each data point in every column is of type `object`. This will cause problems when we want to do statistics with our data later on. In order to perform mathematical operations on our data, our data must be of the correct data type. 

To solve this, we will re-structure our data frame by splitting the column `School Rankings` into two separate columns, one with just the number (e.g., 5) and one with just the text (e.g., 'Extreme'). 

```python
school_rankings = []
for ranking in df['School Rankings']:
    school_ranking = ranking.split('-')
    school_ranking = (int(school_ranking[0].strip()), school_ranking[1].strip())
    school_rankings.append(school_ranking)

df['School Rankings #'] = [ranking[0] for ranking in school_rankings]
df['School Rankings Text'] = [ranking[1] for ranking in school_rankings]
df.head()
```

From the above, you can see that we have two new columns. Let's get the info for this reformatted dataframe. 

```python
df.info()
```

We can now see that we have 8 columns (instead of 6), as we have added 2 more columns to our dataframe. Furthermore, we can see that the column `School Rankings #` is of data type `int64`, the data type we need to perform mathematical operations. In fact, we want all columns with numerical data like `# Faculty Boycotters` and `FJP Events & Statements` to be of type `int54`, so let's convert these columns now, and inspect our dataframe info once more. 

```python
df['# Faculty Boycotters'] = df['# Faculty Boycotters'].astype(int)
df['FJP Events & Statements'] = df['FJP Events & Statements'].astype(int)

df.info()
```

Our data is now ready to do statistics. 

```python
df.describe()
```

## Data analysis 

One thing we might be interested in is the number of schools that have each ranking (0-5). We display this information below.

```python
ranking_counts = df['School Rankings #'].value_counts().sort_index()
print(ranking_counts)
```

As we can see, there are 485 schools with a ranking of 0 ('Negligible'), 42 with a ranking of 1 ('Minimal'), and so on and so forth. 

We can also visualize this information in a bar chart.

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(ranking_counts.index, ranking_counts.values)
plt.xlabel('Ranking')
plt.ylabel('Count')
plt.title('School Rankings')
plt.show()
```

This looks nice. However, we might want the colors of each bar in the chart to match the colors used on the actual webpage for each ranking, e.g., red for 5, green for 0, etc.

We will obtain this information from the HTML as before. We are interested in all data points that have styling information, in particular ones with a `background-color`.

```python
td_tags = soup.find_all('td', style=True)
td_tags[1]
```

We can extract all the background colors we need with the following code below: 

```python
background_colors = []
for td in td_tags:
    styles = td['style']
    styles = styles.split(';')
    for style in styles:
        if 'background-color' in style:
            background_color = style.split(':')[1]
            if background_color not in background_colors:
                background_colors.append(background_color)

background_colors.reverse()
print(background_colors)
```

We now have a list of all hex color codes used in the table of our webpage. We can use this list of `background_colors` to re-create our bar chart. 

```python
plt.bar(ranking_counts.index, ranking_counts.values, color=background_colors)
plt.xlabel('Ranking')
plt.ylabel('Count')
plt.title('School Rankings')
plt.show()
```

We can also display our data with the highest (worst) rank on the left, and the lowest (best) rank on the right. This feels more in line with the order of data on the actual webpage, with higher (worst) ranks at the top and lower (best) ranks at the bottom. 

```python
plt.bar(ranking_counts.index, ranking_counts.values, color=background_colors)
plt.gca().invert_xaxis()  # Reverse the x-axis
plt.xlabel('Ranking')
plt.ylabel('Count')
plt.title('School Rankings')
plt.show()
```

From this chart, we can see the distribution of schools organized by school ranking. It appears that the majority of schools have a favorable 0/"Negligible" ranking.

We might also be interested in exploring the presence of anti-Zionist faculty in a particular state, e.g., Florida. A simple way we can do this is by extracting all schools with the word 'Florida' (for instance) from our dataframe. To do so, we create a new dataframe `florida` with the same column header names, but this time we only add rows with the word 'Florida' in column `College/University`. We do this below, and then inspect the dataframe. 

```python
florida = pd.DataFrame(columns=df.columns)
florida
```

```python
for index, row in df.iterrows():
    if 'Florida' in row['College/University']:
        length = len(florida)
        florida.loc[length] = row
```

```python
florida_ranking_counts = florida['School Rankings #'].value_counts().sort_index()
all_ranks = range(0,6)  # Full range of ranks
florida_ranking_counts = florida_ranking_counts.reindex(all_ranks, fill_value=0)  # Fill missing ranks with 0
florida_ranking_counts
```

```python
plt.bar(florida_ranking_counts.index, florida_ranking_counts.values, color=background_colors)
plt.gca().invert_xaxis()  # Reverse the x-axis
plt.xlabel('Ranking')
plt.ylabel('Count')
plt.title('Florida School Rankings')
plt.show()
```

As we can see, the majority of schools in Florida have a favorable rating, with only 2 schools with an unfavorable rating. 

Looks like Florida has a good grip on its antisemitism.
